```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
qb_data = pd.read_csv('../data/final/qb_train.csv')
qb_data = qb_data[qb_data.year > 2010]
```

```python
qb_test_data = pd.read_csv('../data/final/qb_test.csv')
```

```python
preds = ['pass yds', 'cmp %', 'td', 'int', '40+', 'sck']
target = ['salary_per_year']

X = qb_data[preds]
y = qb_data[target]
```

```python
import statsmodels.api as sm

X= sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_test = qb_test_data[preds]
X_test = sm.add_constant(X_test)
y_test = qb_test_data[target]

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R² Score:", r2)
```

```python
qb_test_data['pred'] = y_pred
qb_test_data.sample(6)
```
