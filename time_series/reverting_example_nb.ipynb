```python
from time_series import revert
import pandas as pd
```

```python
#read in prediction data, here I am just using adjusted salaries as an example
data = pd.read_csv('QB_adjusted_salaries.csv')
data = data[data['year_signed']==2024]
data['predicted_salary'] = data['med_adjusted_apy']
```

```python
reverting = revert(data, position = 'QB', year = 2024)
```

```python
#reverting.unnormalize('median')
```

```python
#2025 example, important to realize we are using 2024 scalers since we dont have full 2025 data
data = pd.read_csv('QB_adjusted_salaries.csv')
data = data[data['year_signed']==2025]
data['predicted_salary'] = data['smooth_adjusted_apy']
```

```python
reverting = revert(data, position = 'QB', year = 2025)
```

```python
reverting.unnormalize('smooth')
```

```python
#mean example, uses inverse transform method from sklearn standard scaler class
data = pd.read_csv('QB_adjusted_salaries.csv')
data = data[data['year_signed']==2025]
data['predicted_salary'] = data['mean_adjusted_apy']
```

```python
reverting = revert(data, position = 'QB', year = 2025)
```

```python
reverting.unnormalize('mean')
```
